```python
import openai
import random
import json

# Load slang dictionary
SLANG_DICT = {
    "normal": {},  # No slang, just casual speech
    "slang": {
        "hello": "yo", "what's up": "wassup", "goodbye": "peace",
        "friend": "homie", "great": "lit", "understand": "get it",
        "money": "cash", "food": "grub", "cool": "fire"
    },
    "ultra_genz": {
        "hello": "yoo wassup", "what's up": "wyd bro", "goodbye": "ight bet",
        "friend": "bestie", "great": "lowkey fire", "understand": "big brain",
        "money": "bag", "food": "snack", "cool": "drippy"
    }
}

def apply_slang(text, level):
    words = text.split()
    for i, word in enumerate(words):
        for key, value in SLANG_DICT.get(level, {}).items():
            if word.lower() == key:
                words[i] = value
    return " ".join(words)

import openai

def chat_with_ai(user_input, api_key, slang_level="normal"):
    client = openai.OpenAI(api_key=api_key)  # ✅ Use OpenAI client object
    response = client.chat.completions.create(  # ✅ New syntax
        model="gpt-3.5-turbo",
        messages=[
            {"role": "system", "content": "You are a high schooler who talks like today's teens."},
            {"role": "user", "content": user_input}
        ]
    )
    
    ai_response = response.choices[0].message.content
    return apply_slang(ai_response, slang_level)


def main():
    api_key = input("Enter your OpenAI API key: ")
    slang_level = input("Choose slang level (normal, slang, ultra_genz): ").strip().lower()
    if slang_level not in SLANG_DICT:
        slang_level = "normal"
    
    print("\nChatbot is ready! Type 'exit' to quit.")
    while True:
        user_input = input("You: ")
        if user_input.lower() == "exit":
            print("Chatbot: Aight, peace ✌️")
            break
        response = chat_with_ai(user_input, api_key, slang_level)
        print(f"Chatbot: {response}")

if __name__ == "__main__":
    main()
```
